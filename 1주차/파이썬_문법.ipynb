#### print() 관련 에러 살펴보기

`print`의 괄호를 제대로 닫지 않았을 때 `SyntaxError`가 발생합니다.   
에러 내용을 보면 `)`가 닫히지 않았다고 합니다.   
에러가 뜨면 당황하지 마시고 에러를 찬찬히 읽어보면 에러를 해결하는데 힌트를 얻을 수 있습니다.

python에서는 괄호를 열어주는 부분과 닫아주는 부분의 짝을 추가해줍시다.

```python
print("3"
# print("3")
```

괄호를 닫아주는 부분과 매치가 되는 괄호를 여는 부분이 없다 합니다.   
괄호를 여는 부분을 추가해줍시다.

```python
print"3")
# print("3")
```

# 8. 반복적이고 따분한 작업을 떠넘기기
인간이 지금까지 진화하며 살아남은 이유는 도구를 잘 쓰기 때문입니다

CPU는 사람의 암산능력과 비할바 안되는 굉장히 빠른 연산속도를 가지고 있습니다

우리는 이 CPU를 통해 반복적인 업무를 처리함으로서 한걸음 더 진화해볼겁니다

```python
nums = [10, 20, 30, 40, 50]
for num in nums:
    num = num + 3
    print(num)
print("for문 끝!")
```

```python
def get_BMI(weight, height):
    height = height / 100 
    bmi = weight / (height ** 2)
    return bmi

bmi = get_BMI(71.5, 176.5)
bmi
```

```python
print(get_BMI(91.5, 182.5))
print(get_BMI(81.5, 189.5))
print(get_BMI(41.5, 155.5))
print(get_BMI(51.5, 164.5))
```
