```python
import plotly.express as px
```

```python
import pandas as pd
flights_df = pd.read_csv("flights.csv")
```

#### 연도별로 각 월 탑승객 수 변화 라인그래프 그리기

```python
fig = px.line(
    flights_df,
    x="month",
    y="passengers",
    color="year",
)
fig
```

#### 그래프의 제목 설정하기

```python
fig = px.line(
    flights_df,
    x="month",
    y="passengers",
    color="year",
)
fig.update_layout(
    title="Passengers per month by year",
    title_x = 0.5,
)
```
