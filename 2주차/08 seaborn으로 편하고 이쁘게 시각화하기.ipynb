```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
flights_df = pd.read_csv("flights.csv")
```

```python
sns.lineplot(
    data=flights_df[flights_df["year"]==1949],
    x="month",
    y="passengers"
)
plt.show()
```

#### 1949년의 월별 비행횟수를 바그래프로 그리기

```python
sns.barplot(
    data=flights_df[flights_df["year"]==1949],
    x="month",
    y="passengers"
)
plt.show()
```

#### 연도별 탑승객 수 박스차트 그리기

```python
plt.figure(figsize=(10, 4))
sns.boxplot(
    data=flights_df,
    x="year",
    y="passengers"
)
plt.show()
```
