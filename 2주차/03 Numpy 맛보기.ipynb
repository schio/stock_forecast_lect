# Numpy 맛보기

#### Numpy의 간단한 사용법

```python
import numpy as np
```

```python
my_list = [10, 20, 30, 40, 50]
print(my_list)
print(type(my_list))
```

```python
my_array = np.array(my_list)
print(my_array)
print(type(my_array))
```

#### array로 통계값 계산하기

```python
print(my_array.min())
print(my_array.max())
print(my_array.mean())
print(my_array.std())
```

#### numpy를 직접적으로 사용하여 통계값 계산하기

```python
print(np.min(my_array))
print(np.max(my_array))
print(np.mean(my_array))
print(np.std(my_array))
```
