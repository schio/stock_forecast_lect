```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
flights_df = pd.read_csv("flights.csv")
flights_df
```

```python
flights_1949_df = flights_df[flights_df["year"]==1949]
plt.plot(
    flights_1949_df["month"],
    flights_1949_df["passengers"]
)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    flights_1949_df["month"],
    flights_1949_df["passengers"]
)
plt.show()
```

#### 그래프의 제목 설정하기

```python
plt.figure(figsize=(10, 6))
plt.plot(
    flights_1949_df["month"],
    flights_1949_df["passengers"]
)
plt.title("Passengers in 1949")
plt.show()
```
